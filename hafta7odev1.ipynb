```python
genre = data["genre"].value_counts()
genre
```

```python
plt.figure(figsize=(18,5))
sns.barplot(x=genre.index , y=genre.values)
plt.xticks(rotation=90);
plt.xlabel("genre")
plt.title("Genres List")
```

```python
data
```

```python
totalgross_genre = data.groupby("genre")[["total_gross"]].mean().reset_index()
totalgross_genre
```

```python
plt.figure(figsize=(18,5))
sns.barplot(x="genre" , y="total_gross", data=totalgross_genre)
plt.xticks(rotation=90);
plt.xlabel("genre")
plt.ylabel("gross")
plt.title("gross by genre")
```

```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import pandas as pd
```

```python
fig = make_subplots(x_title = "release_date",y_title = "Gross",rows = 2, cols= 1)

fig.add_trace(go.Scatter(x = data["release_date"], y = data["total_gross"], name="total_gross"), row = 1, col = 1)

fig.add_trace(go.Scatter(x = data["release_date"], y = data["inflation_adjusted_gross"], name="inflation_adjusted_gross"), row = 2, col = 1)

fig.update_yaxes(range=[0,13000], row=3, col=1)
fig.update_layout(title_text = "Gross by Date")
fig.show()
```
